```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# data sets  library
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
X , y_true = make_blobs(n_samples=500, centers=3, cluster_std=0.60, random_state=42)
```

```python
df = pd.DataFrame(X)
```

```python
df
```

- chuki ye data distance based me hai to value ke karan jyada distance n aye 
iske liye  standardization karna padega

```python
sclaer = StandardScaler()
X_scaled = sclaer.fit_transform(X)
```

```python
# wcss = inertia 
inertia = [] 
k_range = range(1,11)
```

```python
for k in k_range:
    kmean = KMeans(n_clusters=k , random_state=42)
    kmean.fit(X_scaled)
    inertia.append(kmean.inertia_)
```

```python
inertia
```

```python
plt.plot(k_range , inertia, marker='o')
```

```python
kmeans_final = KMeans(n_clusters=3, random_state=42)
```

```python
cluster_lables = kmeans_final.fit_predict(X_scaled)
```

```python
df['cluster'] = cluster_lables
```

```python
cluster_lables
```

```python
print(df.columns)
```

```python
sns.scatterplot(
    x=df[0],
    y=df[1],
    hue=df['cluster'],
    palette='viridis'
)
```

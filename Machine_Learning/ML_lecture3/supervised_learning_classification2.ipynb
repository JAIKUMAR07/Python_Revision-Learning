## 1. What is a Decision Tree?

A decision tree is a supervised learning algorithm used for classification and regression.

It works like a flowchart:

**Internal nodes** ‚Üí represent conditions (tests on features).

**Branches** ‚Üí represent outcomes of conditions.

**Leaf nodes** ‚Üí represent final decision/output (class label or value).

1. What is SVM?

SVM (Support Vector Machine) is a supervised machine learning algorithm primarily used for classification, but it can also be adapted for regression. The main idea is to find a decision boundary (hyperplane) that best separates data points of different classes.

In simple terms: it tries to draw the best line (or hyperplane) between classes so that they are separated as clearly as possible.

2. Key Concepts of SVM
a. Hyperplane

A hyperplane is a decision boundary that separates different classes.

In 2D, it‚Äôs a line. In 3D, it‚Äôs a plane. In higher dimensions, it‚Äôs a hyperplane.

. Margin

The margin is the distance between the hyperplane and the nearest data points from each class.

SVM goal: Maximize the margin ‚Üí called Maximum Margin Classifier.

Larger margin ‚Üí better generalization on unseen data.

c. Support Vectors

Support vectors are the data points closest to the hyperplane.

They define the position and orientation of the hyperplane.

Only these points matter; the rest of the points do not influence the hyperplane.

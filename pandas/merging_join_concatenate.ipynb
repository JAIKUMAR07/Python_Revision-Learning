```python
import numpy as np
import pandas as pd
```

```python
employess =pd.DataFrame({
'employee_id':[1,2,3,4,5],
'name':['john','emma','Linda','jenny','david'],
'department':['HR','IT','Finance','IT','HR']
})

salaries = pd.DataFrame({
'employee_id':[1,2,3,6,7],
'salary':[5000,6000,4500,7000,5500],
'bonus':[1000,1500,1000,2000,1200]
})
```

```python
employess
```

```python
salaries
```

```python
# by default inner hota hi hai 
pd.merge(employess , salaries, on='employee_id' , how='inner')
```

```python
pd.merge(employess , salaries, on='employee_id' , how='outer')
```

```python
pd.merge(employess , salaries, on='employee_id', how='left')
```

# concatenation 

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                         })

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                         })
```

```python
df1
```

```python
df2
```

```python
# ye concatenation  columns ke based me ho rha 
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2] , axis=1)
```

# Joining 

```python
df1 = pd.DataFrame({
     'name':['Alice','Bob','Charlie'],
}, index=[0,1,2])

df2 = pd.DataFrame({
    'score':[66,77,88],
}, index=[4,1,2])
```

```python
df1.join(df2)  # ye inner join hai 
```

```python
df1.join(df2 , how='outer') 
```

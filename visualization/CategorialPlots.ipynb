# Categorial Data plots 

Categorical plots are used to visualize the relationship between numeric data and categories (like gender, day of the week, or yes/no responses).

# categorical plots :

* box plot
* violin plot
* count plot
* bar plot
* strip plot 
* swarm plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = sns.load_dataset('tips')
df
```

```python
sns.countplot(  df['sex'])
```

```python
sns.countplot( x= df['sex'])
```

```python
sns.countplot( x= df['sex'] , hue=df['smoker'])
```

barplot() requires both x and y parameters

It cannot guess what you want to plot on the y-axis

 isliye x and y dono dena jaruri hai nhi to x bs dene se ajib sa plot show krega 

```python
sns.barplot(x = df['sex'], y = df['total_bill'])

# bar plot me y axis e quantative value   aur x axis me categorial value hote hai 

```

# count plot 

```python
import numpy as np 
sns.barplot(x = df['sex'], y = df['total_bill'] , estimator=np.sum)
```

```python
 
sns.barplot(x = df['sex'], y = df['tip'] , estimator=np.sum)
```

# Box plots 

```python
sns.boxplot( x='tip' , y = 'day', data=df)
```

```python
sns.boxplot( x='tip' , y = 'day', data=df , palette='rainbow' )
```

```python
sns.violinplot(x='day',y='total_bill',data=df,palette='rainbow')
```

# strip plot and swarm plot 

```python
sns.stripplot(x= 'day' ,  y ='total_bill' , data = df)
```

```python
sns.swarmplot(x='day',y='total_bill',data=df)
```

```python
# vilon plot and swarm plot ko combine krna 

sns.violinplot(x='tip' , y ='day' , data=df )
```

```python
sns.violinplot(x='tip' , y ='day' , data=df )
sns.swarmplot(x='tip', y ='day', data=df)
```

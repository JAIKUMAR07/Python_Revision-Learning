we use distribution plots to visualise the distribution of quantitative data 



```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd


```

```python
# online data set hota seaborn ka usko access krna 

df = sns.load_dataset('tips')
df
```

```python
# ye bhi categorigal hua 
df['size'].unique()
```

# histogram plot   



```python
sns.displot(df['total_bill'], bins = 20)

# yaha data log ek continuos bins me occurance jitna hote hai uske according bar bnate 
# ye bins length bhi hum log dikha skte 

# like  1-> 5   ke bich around 10 value aaye to baar ka size 10 jayega 
```

```python
sns.histplot(df['total_bill'], bins = 20 , kde=True)

# KDE kr thoug hhum  skew dkeh lete ki histo gram ka jhukav kaha hai 
```

SNS MAKE FOR GOOD TYPE OF PLOT 

```python
plt.subplot(1,2,1)

sns.histplot(df['total_bill'], bins = 20 , kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'], bins = 20 , kde=True)

 
```

```python
sns.jointplot(x='total_bill',y='tip',data=df ,kind='scatter' )
```

```python
sns.jointplot(x='total_bill',y='tip',data=df ,kind= 'hex'  )
```

```python
sns.pairplot(df)

# each element is realted to another variable 
```

```python
sns.pairplot(df , hue='sex')
# hue ka use krte tb jb do category me difrence dekhna hota 
```

```python
sns.pairplot(df , hue='sex' , palette = 'rainbow')
```

```python
sns.pairplot(df , hue='size' , palette = 'rainbow')
```

```python
sns.rugplot(df['total_bill'])

# yaha wo point show krta  jaha kuch kuch vlaue hota 
```
